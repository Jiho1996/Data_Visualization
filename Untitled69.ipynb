```python
# 서울시 공립 어린이집 or 유치원 건립계획 in 노량진

import csv
import matplotlib.pyplot as plt
import numpy as np
f = open('c:/Users\82107/Desktop/a.csv') 
d= csv.reader(f)
next(d)
tmp = []
tmp2 = []

for k in d :
    if '노량진제1동' in k[0] :                                                 # 노량진1동 찾기
        for i in range (3,11):                                                 # 0세 ~ 7세까지
            tmp.append(int(k[i].replace(',','')))

    if '노량진제2동' in k[0] :                                                 #노량진2동 찾기
        for i in range (3,11):                                                 # 0세 ~ 7세까지
            tmp2.append(int(k[i].replace(',','')))
            
plt.rc('font', family='Malgun Gothic')                                         #한글폰트 사용
plt.xticks(np.arange(0,8,1),('0세','1세','2세','3세','4세','5세','6세','7세')) # x축 지정
plt.plot(tmp, marker='o', label ='노량진제1동')
plt.plot(tmp2,marker='x', label='노량진제2동')
plt.grid()           # 격자생성
plt.xlabel('나이')   # x축제목
plt.ylabel('인구수') #y축제목
plt.legend()         #범례표시
plt.show()
                       
```
