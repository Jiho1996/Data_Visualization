```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(0,50,20)       #random은 마지막50값 포함.
y= np.random.randint(0,50,20)
colors = np.random.randint(0,20,20)  #색 랜덤지정
sizes = np.random.randint(5,20,20)*5 #크기 랜덤지정 *5는 5배크게인듯.

plt.scatter(x,y,c=colors, s=sizes)
plt.show()
```
