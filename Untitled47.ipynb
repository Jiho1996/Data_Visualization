```python
# mathplotlib == 데이터 시각화 ,분석 

import matplotlib.pyplot as plt
import numpy as np 
기온변화 = [23, 30, 33, 27, 30]
기온변화2 = [25, 40, 30, 22, 31]


plt.plot(기온변화, marker='o', color ='g', label = '1994년 기온', markersize=10,linestyle='--',linewidth=5)
plt.plot(기온변화2, marker = '*', color ='r', label ='2019년 기온')

plt.ylim(20,45)              #y축 범위지정
plt.xlim(0,4)                #x축 범위지정

plt.legend(loc = 'upper right')           # 범례생성 (plot에 label이 있어야함)
plt.grid()                   # 격자생성
plt.rc('font', family='Malgun Gothic')    #한글제목용 폰트지정
plt.title('기온')            #제목생성
plt.xlabel('날짜 변화')      #x축 제목
plt.ylabel('최고 기온')      # y축 제목
plt.xticks( np.arange (0,5,1), ('1일', '2일', '3일', '4일','5일'))      # x축 값의 표시단위 붙이기, 끝값 5는 포함하지않는다.

plt.savefig('tmp.png')       #show()전에 호출해야함 저장기능.
plt.show
```
